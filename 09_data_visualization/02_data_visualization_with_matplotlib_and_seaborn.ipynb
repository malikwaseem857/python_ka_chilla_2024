## Data Visualization Guide

## What is Data Visualization?
- Data visualization is the graphical representation of information and data.
- By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.
## Important Libraries for Data Visualization in Python
- **Matplotlib:** Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
- **Seaborn:** Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
-**Plotly:** Plotly is a Python graphing library that makes interactive, publication-quality graphs online.
- **Bokeh:** Bokeh is an interactive visualization library that targets modern web browsers for presentation.
-**Altair:** Altair is a declarative statistical visualization library for Python, based on Vega and Vega-Lite.
- **ggplot:** ggplot is a plotting system for Python based on R's ggplot2 and the Grammar of Graphics. It is built for making profressional looking, plots quickly with minimal code.
- **Pygal:** Pygal is a Python SVG charts creator that takes the hassle out of creating charts for the web.
- **Geoplotlib:** Geoplotlib is a toolbox for creating maps and plotting geographical data.
- **Gleam:** Gleam is inspired by R's Shiny package. It allows users to turn analysis scripts into interactive web apps without needing to know web development.
- **Missingno:** Missingno provides a small toolset of flexible and easy-to-use missing data visualizations and utilities that allows you to get a quick visual summary of the completeness (or lack thereof) of your dataset.
- **Leather:** Leather is the Python charting library for those who need charts now and don’t care if they’re perfect.
Wordcloud: Wordcloud is a tool for making beautiful word clouds.
- **Venn:** Venn is a Python library for creating area-proportional two- and three-way Venn diagrams.
## Grammer of Graphics for plotting in Python
- **Data:** The data to be visualized.
- **Aesthetics:** The visual properties of the data points, such as position, size, shape, color, and transparency.
- **Geometry:** The geometric shapes used to represent the data points, such as bars, lines, and points.
- **Facets:** The arrangement of the data into subsets, such as rows, columns, or panels.
- **Statistics**: The statistical transformations applied to the data, such as aggregations, smoothing, and modeling.
- **Scales:** The mapping of data values to visual properties, such as axis limits, tick marks, and color scales.
- **Themes:** The overall visual appearance of the plot, such as colors, fonts, and gridlines.
- **Guides:** The visual elements used to assist in reading the plot, such as legends, axes, and labels.
- **Interactivity:** The interactive features of the plot, such as tooltips, zooming, and panning.
- **Annotations:** The textual and graphical annotations used to highlight specific aspects of the plot.
- **Extensions:** The additional features and plugins that enhance the plot, such as animations, sliders, and widgets.
## Parts of a plot
-** Title:** The title of the plot, which describes the main purpose or content of the visualization.
- **Axis Labels:** The labels for the x and y axes, which describe the variables being plotted.
- **Tick Marks:** The marks along the axes that indicate the scale or range of the data.
- **Gridlines:** The lines that extend from the tick marks to assist in reading the plot.
- **Legend:** The key that explains the colors, shapes, or sizes used to represent different groups or variables.
- **Data Points:** The individual points, bars, lines, or other shapes that represent the data values.
- **Data Labels:** The text or numbers that provide additional information about the data points.
- **Annotations:** The additional text or shapes that highlight specific aspects of the plot.
-** Background:** The overall appearance of the plot, including colors, fonts, and borders.

## Data Visualization with Matplotlib and Seaborn

###  Loading the Libraries and Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the sample diamonds data
df = pd.read_csv('../00_datasets/diamonds.csv')

# Display the first few rows of the dataset
df.head()
```

## Explain each part of diamonds dataset:
A data frame with 53,940 rows and 10 variables:

- **carat:** Weight of the diamond (0.2--5.01)
- **cut:** Quality of the cut (Fair, Good, Very Good, Premium, Ideal)
- **color:** Diamond color, from D (best) to J (worst)
- **clarity:** A measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- **depth:** Total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
- **table:** Width of top of diamond relative to widest point (43--95)
- **price:** Price in US dollars (
18,823)
- **x:** Length in mm (0--10.74)
- **y:** Width in mm (0--58.9)
-** z:** Depth in mm (0--31.8)

### Basic Plots using Matplotlib


```python
plt.plot(df['carat'], df['price'])
```

```python
# Line Plot
plt.plot(df['carat'], df['price'])
plt.title('Carat vs Price Line Plot')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter Plot
plt.scatter(df['carat'], df['price'])
plt.title('Carat vs Price Scatter Plot')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# Bar Plot
plt.bar(df['cut'].unique(), df.groupby('cut')['price'].mean())
plt.title('Average Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to display all available Seaborn color palettes
def display_palettes():
    palette_names = sorted(sns.palettes.SEABORN_PALETTES)
    n = len(palette_names)
    
    # Create subplots
    fig, axes = plt.subplots(nrows=n, figsize=(8, 2 * n))
    
    for ax, palette_name in zip(axes, palette_names):
        colors = sns.color_palette(palette_name)
        for i, color in enumerate(colors):
            ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
        ax.set_xlim(0, len(colors))
        ax.set_ylim(0, 1)
        ax.set_title(palette_name, fontsize=10, pad=10)
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()

display_palettes()
```
